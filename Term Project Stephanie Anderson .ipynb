```python
import os
from os import linesep
class Contacts:
    def __init__(self, firstname='', lastname='', phonenumber='', email=''):
        self.__firstname = firstname
        self.__lastname = lastname  
        self.__phonenumber = phonenumber
        self.__email = email
    
    def get_firstname(self):
        return self.__firstname
    
    def get_lastname(self):
        return self.__lastname
    
    def get_phonenumber(self):
        return self.__phonenumber
    
    def get_email(self):
        return self.__email
    
    def set_firstname(self, firstname):
        self.__firstname = firstname
        
    def set_lastname(self, lastname):
        self.__lastname = lastname
        
    def set_phonenumber(self, phonenumber):
        self.__phonenumber = phonenumber
    
    def set_email(self, email):
        self.__email = lastname
        
    def __str__ (self):
        return self.__firstname + ' ' +self.__lastname + ',' + self.__phonenumber + ',' + self.__email 
    
```

```python
class Contact_Library:
    def __init__(self):
        self.__contacts_list = []
        
    def add_item(self, new_item):
        self.__contacts_list.append(new_item)
        
    def get_all_items(self):
        return self.__contacts_list.copy() 
    
    def add_all_items_to_list(self):
         return self.__contacts_list.copy()   
    
    def search_item_by_first_name(self,pattern):  
        result = []                        
    
        for item in self.__contacts_list:
            if pattern.lower() in item.get_firstname().lower():          
                result.append(item)                                 
                                                                   
        return result            
    
    def search_item_by_last_name(self,pattern):  
        result = []                        
    
        for item in self.__contacts_list:
            if pattern.lower() in item.get_lastname().lower():          
                result.append(item)
                
        return result   
    
    def search_item_by_phonenumber(self,pattern):
        result = []
        
        for item in self.__contacts_list:
            if pattern in item.get_phonenumber():
                result.append(item)
                                                       
        return result   
    
    def search_item_by_email(self,pattern):
        result = []
        
        for item in self.__contacts_list:
            if pattern.lower() in item.get_email().lower():
                result.append(item)
                                                       
        return result   
    
    def search_first_and_last(self, firstname, lastname):
        found_name = None 
        for item in self.__contacts_list:
            if(firstname.lower() in item.get_firstname().lower()) and lastname.lower() in item.get_lastname().lower():
                found_name = item
                break 
        return found_name
    
    def edit_item (self, firstname, lastname, new_firstname ='', new_lastname = '', new_phonenumber = '', new_email =''):
        found_name = self.search_first_and_last(firstname,lastname)
        if (new_firstname != ''):
            found_name.set_firstname(new_firstname)
        if (new_lastname!=''):
            found_name.set_lastname(new_lastname)
        if (new_phonenumber!=''):
            found_name.set_phonenumber(new_phonenumber)
        if (new_email!=''):
            found_name.set_email(new_email)
        
    def remove_contact (self, firstname, lastname):
        found_name = self.search_first_and_last(firstname,lastname)
        self.__contacts_list.remove(found_name)
        
contact_folder =Contact_Library()
contact_list = [] #creates contact list

```
